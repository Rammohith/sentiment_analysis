Access the data


```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/suhasmaddali/Twitter-Sentiment-Analysis/refs/heads/main/train.csv' )
df
```

Clean The data using regex

```python
import re
```

```python
cleanedlist = []
for i in df['text']:
    # Convert the value to a string, whether it's a float or already a string
    i = str(i)

    # Now apply your regex cleaning
    cleaned_text = re.sub(r'[^\w\s]', '', i)  # Removes everything except word characters and spaces
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)  # Replaces multiple spaces with a single space
    cleaned_data = cleaned_text.strip()  # Removes leading/trailing whitespace
    cleanedlist.append(cleaned_data)
cleanedlist
```

Tokenizing the dataset

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
# Downloading required NLTK datasets
nltk.download('punkt')  # Downloads tokenizer models for sentence and word tokenization
nltk.download('punkt_tab')  # Optional: Additional support for tokenization
```

```python
cleanedlist[0]
```

```python
word_tokenize(cleanedlist[0])
```

```python
tokens = [word_tokenize(x) for x in cleanedlist]
tokens
```

Remove the stopwords from nltk.corpus to simplify the data

```python
from nltk.corpus import stopwords
# Removing stopwords from the tokenized words
# Stopwords are common words (e.g., "the", "is") that do not contribute much to analysis
stop = stopwords.words('english')  # Fetching the list of English stopwords
```

```python
tokens_stop = []

for i in tokens:
    c1 = [j for j in i if j not in stop]
    tokens_stop.append(c1)

tokens_stop
```

```python
tokens_stop[0]
```

Stemming using PorterStemmer

```python
from nltk.stem import PorterStemmer
pc = PorterStemmer()
```

```python
pc.stem(tokens_stop[0][1])
```

```python
stemmed_data = []
for i in tokens_stop:
    s = [pc.stem(j) for j in i]
    stemmed_data.append(s)
stemmed_data
```

Lemmatization using WordNetLemmatizer

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lm = WordNetLemmatizer()
```

Importing pos tags for lemmatizer

```python
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng' )
```

giving pos tags for each word in the data set which are required for lemmatizer

```python
pos_tokens = [pos_tag(i) for i in tokens_stop]
pos_tokens
```

Creating a function to convert POS tags to wordnet tags

```python
from nltk.corpus import wordnet
def generate_pos_tags(tags):
    if tags.startswith('J'):
        return wordnet.ADJ
    elif tags.startswith('V'):
        return wordnet.VERB
    elif tags.startswith('N'):
        return wordnet.NOUN
    elif tags.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
generate_pos_tags(pos_tokens[0][0][0])
```

Performing Lemmatizatipn

```python
lm.lemmatize(pos_tokens[0][1][0],generate_pos_tags(pos_tokens[0][1][1]))
```

```python
lemmed_data = []
for x in range (len(pos_tokens)):
    lem = [lm.lemmatize(pos_tokens[x][y][0],generate_pos_tags(pos_tokens[x][y][1])) for y in range(len(pos_tokens[x]))]
    lemmed_data.append(lem)
lemmed_data
```

vectorization: Using count vectorizer to convert text to numerical form

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
sen_stem = [" ".join(i) for i in stemmed_data]
sen_stem
```

fitting the vectorized data into the model

```python
X_vec = cv.fit_transform(sen_stem)
```

```python
y = df['sentiment']
```

Splitting the data into training and testing set for better model

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(X_vec,y,test_size=0.2,random_state=2)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(train_x,train_y)
```

Prediction

```python
y_pred = model.predict(test_x)
```

Accuracy of the model

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(test_y,y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

Classification report

```python
print("\nClassification Report:")
print(classification_report(test_y, y_pred))
```
